## 결측치 처리

```python
# 결측치 확인
train[train.isna().any(axis=1)]
```

```python
# 결측치 확인
for c in train.columns:
    display(train[c].isna().value_counts())
```

### 지출

```python
# 지출 분포포
pay_numeric_column_list = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

for c in pay_numeric_column_list:
    display(train[c].describe())
```

```python
# 0이 적절, 결측치 처리 완료
for c in pay_numeric_column_list:
    train[c] = train[c].fillna(0)
    display(train[c].isna().value_counts())
```

### 수면

```python
# 일단 저온 수면중인 사람은 지출이 0원이어야 함
# 반대로 돈 쓴 사람은 수면자가 아님

# 쓴 돈
total_spend = train[pay_numeric_column_list].sum(axis=1)

# 돈을 썼으므로 수면자가 아님
train.loc[(total_spend != 0) & (train["CryoSleep"].isna()), "CryoSleep"] = False

# 돈을 안썼으므로 수면자일 확률이 높음(논리적으로 성립하진 않지만 확률적인 이유와 결측치를 처리하기 위해해)
train.loc[(total_spend == 0) & (train["CryoSleep"].isna()), "CryoSleep"] = True
```

```python
# 별다른 단서를 찾을 수 없으므로 중앙값으로 처리
train["Age"].describe()

train["Age"] = train["Age"].fillna(train["Age"].median())
```

# 학습

## 학습데이터 준비

```python
y = train["Transported"]
X = train.drop(columns="Transported")
X.columns, y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=2026)

display(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

## 랜덤 포레스트

```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=500, 
    random_state=2026
)

rf.fit(X_train, y_train)

print(f"RandomForest 점수: {rf.score(X_val, y_val):.4f}")
```

## XGBoost

```python
from xgboost import XGBClassifier

xgboost = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=3,
    random_state=2026,
    eval_metric='logloss'
)

xgboost.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=False
)

print(f"XGBoost 점수: {xgboost.score(X_val, y_val):.4f}")
```

## LightGBM

```python
import lightgbm as lgb

lgbm_params = {
    "n_estimators": 1000,
    "learning_rate": 0.05,
    "num_leaves": 31,
    "max_depth": -1,
    "random_state": 2026,
    "n_jobs": -1,
}

lgbm = lgb.LGBMClassifier(**lgbm_params)


callbacks = [
    lgb.early_stopping(stopping_rounds=50),
    lgb.log_evaluation(period=20)
]

lgbm.fit(
    X_train,
    y_train,
    eval_set=[(X_val, y_val)],
    eval_metric="binary_logloss",
    callbacks=callbacks
)

val_score = lgbm.score(X_val, y_val)
print(f"LightGBM Validation Score: {val_score:.4f}")
```

## 앙상블

```python
from sklearn.ensemble import VotingClassifier

estimators = [
    ("rf", rf),
    ("xgb", xgboost),
    ("lgbm", lgbm)
]

voting_model = VotingClassifier(
    estimators=estimators,
    voting="soft",
    weights=[1, 2, 3],
    n_jobs=-1
)

voting_model.fit(X_train, y_train)
ensemble_score = voting_model.score(X_val, y_val)

print(f"Ensemble (Soft Voting) Score: {ensemble_score:.4f}")
```

## 그리드 탐색

```python
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV

lgbm = lgb.LGBMClassifier(random_state=2026)

param_grid = {
    "n_estimators": [50, 100],
    "max_depth": [1, 10],
}

grid_model = GridSearchCV(
    estimator=lgbm,
    param_grid=param_grid,
    scoring="accuracy",
    cv=5,
    verbose=1,
    n_jobs=-1
)

grid_model.fit(X_train, y_train)

print(f"Best Parameters: {grid_model.best_params_}")
print(f"Best CV Score: {grid_model.best_score_:.4f}")
```
