```python
# 프로젝트 1 - 사각형 계산기
class Square:
    def __init__(self):
        self.bot_width = 0
        self.top_width = 0
        self.height = 0
        
    def router(self):
        what_square = input("평행사변형(1), 사다리꼴(2), 끝내려면(3)")
        
        return what_square            
        
    def par(self):
        while True:
            try:
                self.bot_width = float(input("가로? "))
                self.height = float(input("세로? "))
                
            except ValueError as e:
                print(f"잘못된 입력입니다. {e}")
                
            else:
                area = self.bot_width * self.height
                result = f"넓이는 {area}"
                break
                
        return result
    
    def trape(self):
        while True:
            try:
                self.bot_width, self.top_width = map(float, input("아랫변과 윗변? (ex 1, 2) ").split(","))
                self.height = float(input("세로? "))
                
            except ValueError as e:
                print(f"잘못된 입력입니다. {e}")
                
            else:
                if self.bot_width > self.top_width:
                    area = (self.bot_width * self.height) - (((self.bot_width - self.top_width) * self.height) / 2)
                    break
                    
                else:
                    area = (self.top_width * self.height) - (((self.top_width - self.bot_width) * self.height) / 2)
                    break
        
        result = f"넓이는 {area}"
        
        return result


def run():
    global square_calc
    square_calc = Square()
    
    while True:
        try:
            router = square_calc.router().lower()
            
            if router == "1":
                print(square_calc.par())
                
            elif router == "2":
                print(square_calc.trape())
                
            elif router == "3":
                print("종료.")
                break
                
            else:
                raise ValueError("잘못된 명령어 입력.")
            
        except ValueError as e:
            print(f"다시 입력해주세요. ({e})")

run()   
    
```

```python
# 프로젝트 2 - 키오스크
import re
import time


class Menu:
    def __init__(self):
        self.name = ["커피(1)", "유자차(2)"]
        self.price = [4500, 3000]
        self.stock = [30, 50]

    def add_menu(self, name: str, price: int):
        self.name += [name + f"({len(self.name) + 1})"]
        self.price += [price]


class Option:
    def __init__(self):
        self.temperature = ["Hot(1)", "Ice(2)", "Moderate(3)"]
        self.cup = ["Mug(1)", "Plastic(2)"]
    
    def add_option(self, temperature: str, cup: str):
        self.temperature += [temperature + f"({len(self.temperature) + 1})"]
        self.cup += [cup + f"({len(self.cup) + 1})"]


def assembly():
    menu = Menu()
    option = Option()
    return menu, option


class Kiosk:
    def __init__(self, menu, option):
        self.name = menu.name
        self.price = menu.price
        self.temperature = option.temperature
        self.cup = option.cup
        self.orders = []
        self.log = []

    def get_orders(self):
        answer = "1"
        while answer == "1":
            try:
                choice = input(f"메뉴 번호를 입력해주세요. {self.name}, 취소(exit)")
                if choice.strip().lower() == "exit":
                    break
                else:
                    ordered_name = self.name[int(choice) - 1]
                    ordered_price = self.price[int(choice) - 1]

                choice = input(f"온도 번호를 입력해주세요. {self.temperature}, 취소(exit)")
                if choice.strip().lower() == "exit":
                    break
                else:
                    ordered_temperature = self.temperature[int(choice) - 1]

                choice = input(f"컵 종류 번호를 입력해주세요. {self.cup}, 취소(exit)")
                if choice.strip().lower() == "exit":
                    break
                else:
                    ordered_take_out = self.cup[int(choice) - 1]

            except ValueError as e:
                self.log += [e, time.strftime('%Y-%m-%d %H:%M:%S')]
                print(f"잘못된 입력입니다. {e}")
            except IndexError as e:
                self.log += [e, time.strftime('%Y-%m-%d %H:%M:%S')]
                print(f"선택 목록에 없는 번호입니다. {e}")
            except Exception as e:
                self.log += [e, time.strftime('%Y-%m-%d %H:%M:%S')]
                print(f"예상치 못한 에러입니다. 처음부터 다시 시작해주세요. {e}")
                break            
            else:
                cleaned_name = re.sub(r"[^a-zA-Z가-힣\s]", "", ordered_name)
                cleaned_temperature = re.sub(r"[^a-zA-Z가-힣\s]", "", ordered_temperature)
                cleaned_take_out = re.sub(r"[^a-zA-Z가-힣\s]", "", ordered_take_out)
                self.orders += [[cleaned_name, ordered_price, cleaned_temperature, cleaned_take_out]]
                
                answer = input("더 주문하시겠습니까? 예(1), 아니오(아무키나 누르세요.)")

    def print_order_sheet(self):
        total_price = 0
        print("---------------------------------------------------------------------")
        print("------------------------------ 주문서 --------------------------------")
        print("---------------------------------------------------------------------")
        print("---------------------------------------------------------------------")
        
        for i in range(len(self.orders)):
            total_price += self.orders[i][1]
            sentence_length = len(f"주문 메뉴: {self.orders[i][0]}, 온도: {self.orders[i][2]}, 컵 종류: {self.orders[i][3]} | 가격: {self.orders[i][1]}")
            print(f"{"-" * int((69 - sentence_length)/3.5)}"+ " " + f"주문 메뉴: {self.orders[i][0]}, 온도: {self.orders[i][2]}, 컵 종류: {self.orders[i][3]} | 가격: {self.orders[i][1]}" + " " + f"{"-" * int((69 - sentence_length)/3.5)}")
            print("---------------------------------------------------------------------")
            
        print("---------------------------------------------------------------------")
        print(f"-------------------------- 총합: {total_price}원 -----------------------------")
        print("---------------------------------------------------------------------")
        print("---------------------------------------------------------------------")
            

# def add_menu_func():
#     parts = assembly()
#     parts[0].add_menu("밀크티", 5000)

# menu = Menu()
# menu.add_menu("밀크티", 5000)
# print(menu.name)

def run():
    menu, option = assembly()
    menu.add_menu("밀크티", 5000)
    kiosk = Kiosk(menu, option)
    kiosk.get_orders()
    kiosk.print_order_sheet()
    print(f"오류 내역: {kiosk.log}")
    
run()
```
